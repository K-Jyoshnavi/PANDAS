```python
all_data['Order Date'] = pd.to_datetime(all_data['Order Date'], format='%m/%d/%y %H:%M',errors='coerce')
all_data['date_info'] = all_data['Order Date'].dt.date
all_data['time_info'] = all_data['Order Date'].dt.time
all_data.head()
```
